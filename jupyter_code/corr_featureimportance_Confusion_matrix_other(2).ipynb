```python
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np

import pandas
plt.figure(figsize=(8.5,7))

# load dataset
dataframe = pandas.read_table('/content/completeallwithNode.txt')
data1 = dataframe.drop("No.", axis=1)
# print(data1)
data1.rename(columns={"State": "Attacked"}, inplace = True)
data1["Attacked"].replace({"T": 0, "F": 1}, inplace=True)

print(data1)


corrMatrix = data1.corr(method = 'pearson')

print(corrMatrix)
plt.rcParams['font.size'] = 10
plt.rcParams['font.weight'] = 'bold'
sn.heatmap(corrMatrix, annot = True, fmt = '.2g')
plt.xticks(rotation=45, ha ='right', fontsize = 15, fontweight = 'bold' )
plt.yticks(rotation=45, ha ='right', fontsize = 15, fontweight = 'bold')


plt.savefig('corr_mixed.PNG', bbox_inches = "tight")
plt.savefig('corr_mixed.svg', bbox_inches = "tight")
plt.savefig('corr_mixed.pdf', bbox_inches = "tight")

print("Hmm")





```
