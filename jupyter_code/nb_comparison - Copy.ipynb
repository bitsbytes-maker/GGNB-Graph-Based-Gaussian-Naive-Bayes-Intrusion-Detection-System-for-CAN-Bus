```
import seaborn as sn
import matplotlib.pyplot as plt
import pandas
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import naive_bayes as nb 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import seaborn as sns
import pandas as pd
from pylab import savefig
import matplotlib.pyplot as plt

def confusion_matrix11(file,num):
# load dataset
  dataframe = pandas.read_table(file)
  data = dataframe.drop("No.", axis=1)
  dataset = data.values

  print(dataset.shape)

  # split into input (X) and output (Y) variables

  #l = 36893 
  l = num
  X = dataset[:,0:]
  # Y = dataset[100:200, 1:10]
  # for col in data.columns: 
  #     print(col)
  # print(X.shape)
  labels = []
  for i in range (0,l):
    if X[i][9] == 'T':
      labels.append(0)
    else:
      labels.append(1)

  features = X[0:l,0:9]

  # print(features)

  # print(features.shape)
  # df = dataframe.corr()

  # #print(df)

  # corrMatrix = data.corr()
  # print(corrMatrix)

  # sn.heatmap(corrMatrix, annot=True, fmt = ".1g")
  # plt.show()




  # Split our data
  train, test, train_labels, test_labels = train_test_split(features,
                                                            labels,
                                                            test_size=0.33,
                                                            random_state=42)#

  # Initialize our classifier
  gnb = nb.ComplementNB()
  

  # Train our classifier
  model = gnb.fit(train, train_labels)

  preds = gnb.predict(test)
  cm = confusion_matrix(test_labels, preds)
  # print(preds)
  print(accuracy_score(test_labels, preds))
  cls = classification_report(test_labels, preds)
  print(cls)
  return cm

  # Evaluate accuracy
  

cm_dos = confusion_matrix11('/content/complete_dos1.txt', 18565+18328)
cm_fuzzy = confusion_matrix11('/content/complete_fuuzy1.txt',18565+19194)
cm_spoof = confusion_matrix11('/content/complete_spoof1.txt', 18565+23108)
cm_replay =confusion_matrix11('/content/complete_replay1.txt',18565+659)
cm_mixed = confusion_matrix11('/content/completeallwithNode.txt',79854)
print("dos")
print(cm_dos)
print("fuzzy")
print(cm_fuzzy)
print("spoof")
print(cm_spoof)
print("replay")
print(cm_replay)
print("mixed")
print(cm_mixed)

print("overall")

cm_overall = cm_dos+cm_fuzzy+cm_spoof+cm_replay+cm_mixed
print(cm_overall)

print("overall report")
# cls = classification_report(test_labels, preds)
# print(cls)
```
