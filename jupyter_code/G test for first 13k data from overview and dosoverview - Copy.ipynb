```
import numpy as np
import random
#Finding hamming distance function
_xormap = {('0', '1'): '1', ('1', '0'): '1', ('1', '1'): '0', ('0', '0'): '0'}
def xor(x, y):
    return (''.join([_xormap[a, b] for a, b in zip(x, y)])).count('0')

#Readin allfingerprintfiles and store it in a list
def stringViewList(filename):

  file_view = open(filename, 'r')
  list_view = []

  while True:


    line2 = file_view.readline()

    if not line2:
      break
    list_view.append(line2.split())

  file_view.close()


  similarity_score = []
  for i in range(0, len(list_view) - 1):

    xor_distance = xor(list_view[i][0], list_view[i+1][0])
    hamming_distance = xor_distance/128   #1-(1-xor_distance/128)
    similarity_score.append(hamming_distance)

  return similarity_score

#Minimizing the file size by cutting huge data and rename the file 
#orginal filename overView_allfingerprint.txt, dosoverView_allfingerprint.txt
list_overview = stringViewList('/content/overView_allfingerprint_13001.txt')
list_dosoverview = stringViewList('/content/dosoverView_allfingerprint_13001.txt')

print(list_overview)
print(list_dosoverview)

oview_total = 0
dview_total = 0



for i in range (0,13000):
  oview_total += list_overview[i]
  dview_total += list_dosoverview[i]
 

total = oview_total + dview_total

con_oview = []
con_dview = []
for i in range (0,13000):
  ex_oview = ((list_overview[i] + list_dosoverview[i]) * oview_total) / total
  ex_dview = ((list_overview[i] + list_dosoverview[i]) * dview_total) / total 

  con_oview.append(list_overview[i] * np.log(list_overview[i]/ex_oview))
  con_dview.append(list_dosoverview[i] * np.log(list_dosoverview[i]/ex_oview))

file = open('/content/output.txt','w')

for i in range (0, 13000):
  file.write(str(con_oview[i]))
  file.write(" ------> ")
  file.write(str(con_dview[i]))
  file.write("\n")



file.write("length of similarity score of overview data: " +str(len(con_oview)) +"\n")
file.write("length of similarity score of dosoverview data: " +str(len(con_dview))+"\n")
file.write("median of similarity score of overview data: " +str(np.median(con_oview))+"\n")
file.write("median of similarity score of dosoverview data: " +str(np.median(con_dview))+"\n")
file.write("mean of similarity score of overview data: " +str(np.mean(con_oview))+"\n")
file.write("mean of similarity score of dosoverview data: " +str(np.mean(con_dview))+"\n")
file.write("variance of similarity score of overview data: " +str(np.var(con_oview))+"\n")
file.write("variance of similarity score of dosoverview data: " +str(np.var(con_dview))+"\n")
file.write("standard deviation of similarity score of overview data: " +str(np.std(con_oview))+"\n")
file.write("standard deviation of similarity score of dosoverview data: " +str(np.std(con_dview)))

file.close()

print("length of similarity score of overview data: " +str(len(con_oview)))
print("length of similarity score of dosoverview data: " +str(len(con_dview)))
print("median of similarity score of overview data: " +str(np.median(con_oview)))
print("median of similarity score of dosoverview data: " +str(np.median(con_dview)))
print("mean of similarity score of overview data: " +str(np.mean(con_oview)))
print("mean of similarity score of dosoverview data: " +str(np.mean(con_dview)))
print("variance of similarity score of overview data: " +str(np.var(con_oview)))
print("variance of similarity score of dosoverview data: " +str(np.var(con_dview)))
print("standard deviation of similarity score of overview data: " +str(np.std(con_oview)))
print("standard deviation of similarity score of dosoverview data: " +str(np.std(con_dview)))

# llll = []
# for i in range (0, 13000):
#   if np.mean(con_oview) + np.std(con_dview) < con_dview[i] or np.mean(con_oview) - np.std(con_dview) > con_dview[i]:

#     llll.append(i)


```
