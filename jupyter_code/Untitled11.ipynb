```
import numpy as np


fileId = open('/content/DoS_dataset.csv', 'r')

detectList = []
counter = 1
graphList = []
ultimateGraphList = [] 
utlimateaDetectionList = []
graphCounter = 0

nodeList = []
messageSize = 200

for line in fileId:

  Xpartition = line.split()[0].split(',')
  node = Xpartition[1]
  nodeList.append(node)
  attackCheck = Xpartition[len(Xpartition)-1]



  graphList.append(node)
  detectList.append(attackCheck)
  

  if len(graphList) == messageSize: 
    ultimateGraphList.append(graphList)
    utlimateaDetectionList.append(detectList)
   

    graphList = []
    detectList = []
   
   

ugl = ultimateGraphList

udl = utlimateaDetectionList

print(len(ugl))
print(len(udl))

print(ugl[0])
print(len(ugl[0]))
print(len(udl[0]))
print(udl[0])
print(udl[3])



nodeList = list(set(nodeList))
print(nodeList)
ln = len(nodeList)
print(ln)
print(nodeList.index('0000'))
```

```
#Making graph from the 200 sequential of id and store them in a file as Graph adjacent list
from collections import defaultdict
import networkx as nx
import matplotlib.pyplot as plt
import random
from scipy.sparse import csr_matrix 
from spektral.data import graph
gAttackfree=nx.DiGraph()
gAttacked = nx.DiGraph()

gGraph = nx.DiGraph()

lenugl = len(ugl)
uList = []
vList = []
uvList = []


counter = 0
gAttackfreeCounter = 0
gAttackedCounter = 0

allGraphList = []

for i in range(0,lenugl):

  dicGraph = defaultdict(list)
  attackFree = 1

 
  uvList = np.zeros((ln,ln))

  

  for j in range (0, len(ugl[i])-1):
    u = nodeList.index(str(ugl[i][j]))
    v = nodeList.index(str(ugl[i][j+1]))
    uvList[u][v] = 1.0
    gGraph.add_edge(nodeList.index(str(ugl[i][j])),nodeList.index(str(ugl[i][j+1])))


  Adjacency = csr_matrix(uvList)
  
  

  pg = nx.pagerank(gGraph)
  outDegList = []
  inDegList = []
  outDeg = gGraph.out_degree

  for key,value in outDeg:
    outDegList.append([key,value])



  inDeg = gGraph.in_degree


  for key,value in inDeg:
    inDegList.append([key,value])


  pagrankList = []
  for keys in pg.keys():
    pagrankList.append([keys,pg[keys]])

  pList = sorted(pagrankList, key=lambda x: (x[0]))
  inList = sorted(inDegList, key=lambda x: (x[0]))
  outList = sorted(outDegList, key=lambda x: (x[0]))
  
  
  # print(pList)
  # print(inList)
  # print(outList)

  nodeFeatures = []

  for j in range(len(pList)):
    nodeFeatures.append([float(pList[j][1]),float(inList[j][1]),float(outList[j][1])])
  
  nf = np.array(nodeFeatures)


  
  
					
  for j in range (0,len(udl[i])):
    if str(udl[i][j]) == 'T': 
      attackFree = 0
      break

  allGraphList.append(graph.Graph(x= nf, a = Adjacency, e=None, y=attackFree))
    
  


print(allGraphList)

#print(len(list(gAttackfree)))

```
