```
def squrroot(x,y,z=0):
  return (x**2+y**2+z**2)**0.5


import matplotlib.pyplot as plt 
from numpy import random
x = random.randint(100, size=(10, 5))
print(x)
a,b,c,d,e = 0,0,0,0,0
front = x[0]
print("sum")


# increasing the value of coeffiecen to maximize and decreasing to minimize
index = 0
for i in range(1,len(x)):
  squr_root_max = squrroot(x[i][0],x[i][1],x[i][4])
  front_root_max = squrroot(front[0],front[1],front[4])
  squr_root_min = squrroot(x[i][2],x[i][3])
  front_root_min = squrroot(front[2],front[3])
  print(str(x[i][0]+x[i][1]+x[i][4])+"---->"+str(x[i][2]+x[i][3]))
  
  if squr_root_max  >= front_root_max and front_root_min >= squr_root_min :
    front = x[i]
    index = i

print('optimal soulituon')
print(front)
print(str(index+1)+" number row is solution")

```

```
from numpy import random
x = random.randint(100, size=(10, 5))


print(x)
front = x[0]

print("taking intial front")
print(front)
# we try to increase the first, second and fifth coulms
# and try to decreas the third and fourth coulms
for i in range(len(x)):
    # we try to increase the first, second and fifth coulms so their summation in sum_max_x and sum_max_f variables
# and try to decreas the third and fourth coulms so their summation in sum_min_x and sum_min_f variables
  sum_max_x = x[i][0]+x[i][1]+x[i][4]
  sum_max_f = front[0]+front[1]+front[4]
  sum_min_x = x[i][2]+x[i][3]
  sum_min_f = front[2]+front[3]
 # for minimum x <= f and for maximum x >= f It is for summation also 

  if ((x[i][0] >= front[0] or x[i][1] >= front[1] or x[i][4] >= front[1]) 
  and (sum_max_x >= sum_max_f) and (x[i][2] <= front[2] or x[i][3] <= front[3]) and (sum_min_x <= sum_min_f) ):
    front = x[i]


print("best fornt")
print(front)
```
