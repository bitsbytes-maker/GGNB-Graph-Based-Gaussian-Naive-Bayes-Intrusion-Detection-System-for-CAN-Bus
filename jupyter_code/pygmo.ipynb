```python
from pygmo import *
udp = zdt(prob_id = 1)
pop = population(prob = udp, size = 10, seed = 3453412)

from matplotlib import pyplot as plt 
pop = population(udp, 100)
ax = plot_non_dominated_fronts(pop.get_f()) 
plt.ylim([0,6]) 
plt.title("ZDT1: random initial population")  
```
