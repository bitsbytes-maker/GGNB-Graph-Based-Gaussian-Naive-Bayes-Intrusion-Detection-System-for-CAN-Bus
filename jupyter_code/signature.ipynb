```
import pandas as pd
import networkx as nx
import hashlib
import numpy as np


file_overview = open('/content/overView.txt', 'r')


count = 0

graphList = []
ultimate_graphList = []  
graphCounter = 0
while True: 
    count += 1
  
    # Get next line from file 
    line = file_overview.readline() 
  
    # if line is empty 
    # end of file is reached 
    if not line: 
        break
    

    
    graph_number = line[0:12]
    
    if graph_number == "Graph Number":
      #print(graph_number+str(graphCounter))
      graphCounter += 1
      line = file_overview.readline()
      line = file_overview.readline()
      line = file_overview.readline()

      if graphCounter >= 2:
        ultimate_graphList.append(graphList)
        graphList = []


    
    list1 = line.split()
    graphList.append(list1)



ultimate_graphList.append(graphList)
    
file_overview.close()

list_All_Feature_Set = []

graphNumber = len(ultimate_graphList)
# Iterating every graph
for i in range(0,graphNumber):

  G = nx.DiGraph()
  graphEdgeNumber = len(ultimate_graphList[i])
  #making the graph
  for j in range(0,graphEdgeNumber):
    vertexConnectionLength = len(ultimate_graphList[i][j])

    for k in range(2,vertexConnectionLength):
      
      
      G.add_edge(ultimate_graphList[i][j][0],ultimate_graphList[i][j][k])
  weighted_Feature_Set = nx.pagerank(G) 
  

  for a, b in G.edges():
    weighted_Feature_Set[str(a) + " " + str(b)] = (1/len(G.edges(a))) * weighted_Feature_Set[a]

  weighted_Feature_Set1 = {}
  for key, value in weighted_Feature_Set.items():
    sha1_hash = hashlib.sha1(str(key).encode() + str(value).encode()).hexdigest()
    binary_hash = bin(int(sha1_hash, 16))[2:]
      #Trim to fixed size of 128
    binary_hash = (binary_hash[:128]) if len(binary_hash) > 128 else binary_hash
      
    weighted_Feature_Set1[str(binary_hash)] = value  
    
  list_All_Feature_Set.append(weighted_Feature_Set1)

print("length of the feauter set:"+str(len(list_All_Feature_Set)))





all_Fingerprints = []

#Loop through each feature set to generate it's fingerprint
counter=0
for feature_Set in list_All_Feature_Set:
  temp_fingerprint = []
    #Build fingerprint

  for key, val in feature_Set.items():
    binary_list = []
        #For each sha1 hash, if digit=1 then add weighted value. If digit=0, then add negative weighted value.

    for d in key:
      if int(d) == 1:
        binary_list.append(val)
      else:
        binary_list.append(-val)
    temp_fingerprint.append(binary_list)    
     

  #Calculate fingerprint by summing all the columns in temp_fingerprint
  fingerprint = [sum(x) for x in zip(*temp_fingerprint)]
   
    #Convert all positive vals to '1' and all negative vals to '0' in fingerprint
  for index, item in enumerate(fingerprint):
    if item > 0:
      fingerprint[index] = 1
    else:
      fingerprint[index] = 0

    
  all_Fingerprints.append(fingerprint)

print("Lenght of the all_Fingerprints:"+str(len(all_Fingerprints)))


file4 = open('/content/overView_allfingerprint.txt', 'w')

len_finger = len(all_Fingerprints)
for i in range(0,len_finger):
  str1 = ""
  for j in range(0,128):
    str1+= str(all_Fingerprints[i][j])
  file4.write(str1)
  str1=""
  file4.write("\n")
  
  

file4.close()

```

```
# finding similarity score using /content/overView_allfingerprint.txt

_xormap = {('0', '1'): '1', ('1', '0'): '1', ('1', '1'): '0', ('0', '0'): '0'}
def xor(x, y):
    return (''.join([_xormap[a, b] for a, b in zip(x, y)])).count('0')

file_overview = open('/content/overView_allfingerprint.txt', 'r')
list_overview = []

while True:


  line2 = file_overview.readline()

  if not line2:
    break
  list_overview.append(line2.split())

file_overview.close()


file_similarity = open('/content/Oview_similarity.txt', 'w')




len_list_overview = len(list_overview)

similarity_scores_overview = []



for i in range(0,len_list_overview - 1):

  over_string1 = list_overview[i][0]
  over_string2 = list_overview[i+1][0]
  hamming_distance = xor(over_string1,over_string2)
  
  ham_str = (1-(hamming_distance/128))
  similarity_scores_overview.append(ham_str)
  file_similarity.write(str(ham_str))
  file_similarity.write("\n")
    

file_similarity.close()




len_similarity_score = len(similarity_scores_overview)

print("length of the similarity score: "+str(len_similarity_score))





```
